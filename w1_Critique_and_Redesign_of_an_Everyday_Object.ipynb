# **2. Propose a Redesign**

Design Rationale

*   The redesign focuses on the lack of notification, missing temperature, burn risk, and confusing mode selection. They are important because the kettle is frequently used, where mistakes can lead to frustrations and wasting time and energy. With the improvement, it can be easier and safer to use without changing the function.

Proposed Features

*   Sensor (temperature sensor - monitor & display, water-level sensor - detect dry/low water conditions, clearly displayed mode labels, insulated handle & lid - improve grip & burn protection, speaker/buzzer for audible warning notifications)
*   Simple logic system - temperature threshold, safety cutoff, mode control
*   Adding ML might increase cost and reliability


Social Considerations

*   Does not replace existing habits (fill the kettle, press a button, pour out water). The redesign aims to make the smart features invisible, using subtle sounds and a small display, rather than requiring connections to phones. A low social cost - no collection of data, doesnâ€™t need to explain to the user how to use it, and is like a conventional appliance. In general, this would be a non-performative design.


Tradeoff Analysis

*   The redesign will have a small increase in manufacturing cost, components, and possible failure with the screen, sensors, and buzzers. There might also be a risk for users to ignore the alerts or misinterpret the information. But I think these risks are outweighed by improvements in safety and clarity. By keeping it simple, the kettle can limit failure modes and allow it to fall back to the basic boiling and cooking function when advanced features fail.
